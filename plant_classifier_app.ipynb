```python
import pathlib
from pathlib import Path
import platform

from fastai import *
#from fastbook import *
from fastai.vision.widgets import *
from fastai.learner import load_learner
from fastai.vision.core import PILImage
```

```python
plt = platform.system()
if plt == 'Windows':
    pathlib.PosixPath = pathlib.WindowsPath
```

```python
def get_taxa_label(taxa,level='group'):
    levels = ['species', 'genus', 'family', 'group']
    return taxa.parents[levels.index(level)].name
```

```python
path = Path()
learn_inf = load_learner(path/'bing_data_n_750_cleaned_resnet_34.pkl')
```

```python
btn_upload = widgets.FileUpload()
btn_run = widgets.Button(description='Classify')
out_pl = widgets.Output()
lbl_pred = widgets.Label()


def on_click_classify(change):
    img = PILImage.create(btn_upload.data[-1])
    out_pl.clear_output()
    with out_pl: display(img.to_thumb(128,128))
    #lbl_pred.value = 'Test'
    pred,pred_idx,probs = learn_inf.predict(img)
    lbl_pred.value = f'Prediction: {pred}; Probability: {probs[pred_idx]:.04f}'

btn_run.on_click(on_click_classify)
```

```python
VBox([widgets.Label('Upload an image of plant!'), 
      btn_upload, btn_run, out_pl, lbl_pred])
```
